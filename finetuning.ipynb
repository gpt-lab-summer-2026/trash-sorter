```python
import os
os.environ["PYTORCH_ENABLE_MPS_FALLBACK"] = "1"
import torch
from ultralytics import YOLO

print(torch.backends.mps.is_available())
```

```python
model = YOLO("yolov8n-cls.pt")

model.train(
    data="data",
    epochs=50,
    imgsz=224,
    batch=16,
    device="mps",
    name="trashnet_run"
)
```

```python
model = YOLO("runs/classify/trashnet_run/weights/best.pt")
model.export(format="onnx")
```
